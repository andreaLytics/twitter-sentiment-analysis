**Cargar archivo Sentimientos.txt**

```python
sentimientos = open("Sentimientos.txt") 
valores = {} 
for linea in sentimientos: 
    termino, valor = linea.split("\t") 
    valores[termino] = int(valor)
print (valores.items())
```

**Procesamiento de los tweets y cálculo de los sentimientos con la versión 1**

```python
import json
with open("Tweets.txt","r") as Tweets:
    for linea in Tweets:
        tweet = json.loads(linea)
        if 'text' in tweet:
            texto_tweet = tweet['text']
            palabras = texto_tweet.split()
            sentimientos_tweet = 0
            for palabra in palabras:
                sentimientos_tweet += valores.get(palabra.lower(),0)
            #aquí empieza el código para la versión 1, versión 2 en la siguiente celda
            for palabra in palabras:
                palabra_lower = palabra.lower()
                if palabra_lower not in valores:
                    print("{}: {}".format(palabra, sentimientos_tweet))
        else:
                    continue
```

**Cálculo de los sentimientos con la versión 2**

```python
import json
with open("Tweets.txt","r") as Tweets:
    for linea in Tweets:
        tweet = json.loads(linea)
        if 'text' in tweet:
            texto_tweet = tweet['text']
            palabras = texto_tweet.split()
            sentimientos_tweet = 0
            for palabra in palabras:
                sentimientos_tweet += valores.get(palabra.lower(),0)
            #comienza la versión 2
            if sentimientos_tweet > 0:
                media = len(palabras)/sentimientos_tweet
            else:
                media = 0
            for palabra in palabras:
                palabra_lower = palabra.lower()
                if palabra_lower not in valores:
                    print("{}: {}".format(palabra, media))
        else:
                    continue
            
```
